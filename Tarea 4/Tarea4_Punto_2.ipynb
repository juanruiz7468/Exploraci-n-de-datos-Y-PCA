```python
def prediction(W,X):
    resultado=[]
    b=W[0]
    for i in range(0,W.shape[0]-1):
        resultado.append(W[i+1]*X[:,i])
    return b+sum(resultado)
```

```python
def sigmoid(t):
    return 1/(1+np.exp(-t))
```

```python
def label(W,X):
    prediccion=[]
    probabilidad=sigmoid(prediction(W,X))
    for numero in probabilidad:
        prediccion.append(int(round(numero)))
    return prediccion
```

```python
def loss(W,X,Y):
    probabilidad=sigmoid(prediction(W,X))
    for i in range(0,len(probabilidad)):
        loss=(-1*Y[i]*math.log(probabilidad[i],10))-((1-Y[i])*math.log(1-probabilidad[i],10))
    return np.mean(loss)
```

```python
def dloss_dw(W,X,Y,h,batch):
    for b in range(batch):
        aumento_finito=loss(W+h,X,Y)
        diferencia_finita=aumento_finito-loss(W,X,Y)
        dloss_dw=diferencia_finita/h
        h=h/10
    return dloss_dw
```

```python
def gradiente(W,X,Y,h,batch,lr):
    W=W-lr*dloss_dw(W,X,Y,h,batch)
    return W
```

```python
def model(W,X,Y,h,batch,lr,epochs):
    for epoch in range(epochs):
        W=gradiente(W,X,Y,h,batch,lr)
    return W
```
