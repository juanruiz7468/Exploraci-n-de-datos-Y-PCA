```python
# Algoritmo K-Means
import random
import numpy as np

# Se inicializa el contador
i = 0

# Lista de valores de k
k_list = np.arange(1, 5)

# Bucle while con una condición de salida específica
while i < 2:  # Se ejecuta solo una vez
    if i == 0:  # Se ejecuta solo en la primera iteración
        # Iteración sobre las columnas del DataFrame (suponiendo que existe nombres_columna)
        for columna in nombres_columna:
            if columna != "DEPARTAMENTO":  # Asegúrate de que no sea la columna de etiquetas
                for k in k_list:
                    # Inicializa las columnas de centroides para cada columna numérica
                    df["centroide" + str(k) + "_" + columna] = np.nan

        # Genera valores aleatorios para la columna "Cluster_Random0"
        for indice, fila in df.iterrows():
            df.loc[indice, "Cluster_Random0"] = random.randint(1, 4)

        # Cálculo de centroides iniciales
        for k in k_list:
            for columna in nombres_columna:
                if columna != "DEPARTAMENTO":
                    registros_k = 0
                    suma_k = 0
                    
                    for indice, fila in df.iterrows():
                        if df.loc[indice, "Cluster_Random0"] == k:
                            registros_k += 1
                            suma_k += df.loc[indice, columna]
                    
                    if registros_k != 0:
                        valor_centroide = suma_k / registros_k
                        df.loc[0, "centroide" + str(k) + "_" + columna] = valor_centroide

        # Asignación de puntos a centroides más cercanos
        for indice, fila in df.iterrows():
            for k in k_list:
                dist_list = []
                for columna in nombres_columna:
                    if columna != "DEPARTAMENTO":
                        dist = (fila[columna] - df.loc[0, "centroide" + str(k) + "_" + columna])**2
                        dist_list.append(dist)
                
                # Calcula la distancia euclidiana sumando las distancias al cuadrado y luego tomando la raíz cuadrada
                dist_eucli = sum(dist_list)**0.5
                
                # Asigna el punto al centroide más cercano
                df.loc[indice, "distancia" + "_" + str(k)] = dist_eucli
        for indice,fila in df.iterrows():
            min_list=[]
            for k in k_list:
                min_list.append(df.loc[indice,"distancia" + "_" + str(k)])
            centroide_cercano=min_list.index(min(min_list))+1
            df.loc[indice,"clasificacion0"]=centroide_cercano
            
                
    if i > 0:  # Se ejecuta a partir de la segunda iteración
        # Actualización de centroides
        for k in k_list:
            for columna in nombres_columna:
                if columna != "DEPARTAMENTO":
                    # Calcula el valor medio de los registros asociados al cluster k-1 de la iteración anterior
                    valor_centroide = df[df["clasificacion"+str(i-1)] == k][columna].mean()
                    df.loc[0, "centroide" + str(k) + "_" + columna] = valor_centroide
        for indice, fila in df.iterrows():
            for k in k_list:
                dist_list = []
                for columna in nombres_columna:
                    if columna != "DEPARTAMENTO":
                        dist = (fila[columna] - df.loc[0, "centroide" + str(k) + "_" + columna])**2
                        dist_list.append(dist)
                
                # Calcula la distancia euclidiana sumando las distancias al cuadrado y luego tomando la raíz cuadrada
                centroide_cercano = sum(dist_list)**0.5
                
                # Asigna el punto al centroide más cercano
                df.loc[indice, "distancia" + "_" + str(k)] = centroide_cercano
        for indice,fila in df.iterrows():
            min_list=[]
            for k in k_list:
                min_list.append(df.loc[indice,"distancia" + "_" + str(k)])
            centroide_cercano=min_list.index(min(min_list))+1
            df.loc[indice,"clasificacion"+str(i)]=centroide_cercano

    i += 1

print(df.head())
# Suponiendo que 'df' es tu DataFrame y 'carpeta' es la ruta de la carpeta donde deseas guardar el archivo Excel
carpeta = "D:\Esatdistica 3"

# Nombre del archivo Excel
nombre_archivo = "resultado_kmeans.xlsx"

# Ruta completa del archivo Excel
ruta_excel = carpeta + "/" + nombre_archivo

# Guardar el DataFrame como un archivo Excel
df.to_excel(ruta_excel, index=False)

print("Archivo guardado con éxito en:", ruta_excel)
```

```python
from scipy.optimize import minimize
#Algoritmo PCA
df_sin_departamento = df.drop(columns=['DEPARTAMENTO'])
media=df_sin_departamento.mean()
std=df_sin_departamento.std()
df_normalizado=(df_sin_departamento-media)/std
df_normalizado.head()
# Obtener los nombres de las columnas normalizadas
nombres_normalizado = df_normalizado.columns.tolist()

# Crear una matriz de ceros para la matriz de covarianza
matriz_covarianza = np.zeros((len(nombres_normalizado), len(nombres_normalizado)))

# Calcular la matriz de covarianza
for fila in range(len(nombres_normalizado)):
    for columna in range(len(nombres_normalizado)):
        if fila == columna:
            # Varianza de la variable en la diagonal principal
            matriz_covarianza[fila][columna] = df_normalizado[nombres_normalizado[fila]].var()
        else:
            # Covarianza entre las variables
            matriz_covarianza[fila][columna] = df_normalizado[nombres_normalizado[fila]].cov(df_normalizado[nombres_normalizado[columna]])
matriz_identidad=np.eye(3)
# Calcular el valor de lambda utilizando el método de Newton-Raphson
def funcion_lambda(landa):
    return np.linalg.det(matriz_covarianza - (landa * np.eye(len(nombres_normalizado))))

def derivada_lambda(landa):
    return (funcion_lambda(landa + 1e-5) - funcion_lambda(landa)) / 1e-5

# Implementar el método de Newton-Raphson para encontrar la raíz de la función
def Metodo_De_Newton_Rhapson_1(x, max_iteraciones, tolerancia):
    iteraciones = 0
    xn = x
    while iteraciones <= max_iteraciones:
        if derivada_lambda(xn) == 0:
            raise ValueError("La derivada de la función evaluada en ese punto da cero. Intenta cambiar de punto.")
        xn_1 = xn - (funcion_lambda(xn) / derivada_lambda(xn))
        rel_error = abs((xn_1 - xn) / xn_1) if xn_1 != 0 else abs(xn_1 - xn)
        if rel_error < tolerancia:
            return xn_1
        xn = xn_1
        iteraciones += 1
    raise ValueError("El método de Newton-Raphson no convergió después de alcanzar el máximo de iteraciones.")

# Calcular el valor de lambda utilizando el método de Newton-Raphson
landa_1 = Metodo_De_Newton_Rhapson_1(x=1.162, max_iteraciones=100, tolerancia=1e-8)
landa_2 = Metodo_De_Newton_Rhapson_1(x=2.3111, max_iteraciones=100, tolerancia=1e-8)
landa_3 = Metodo_De_Newton_Rhapson_1(x=0.0001, max_iteraciones=100, tolerancia=1e-8)

print("Valores de lambda encontrados:")
print("lambda_1 =", landa_1)
print("lambda_2 =", landa_2)
print("lambda_3 =", landa_3)

# Calculate eigenvalues and eigenvectors of the covariance matrix
eigenvalues, eigenvectors = np.linalg.eig(matriz_covarianza)

# Almacena los valores de lambda en una lista
landas = [landa_1, landa_2, landa_3]

# Calcula los eigenvectores y proyecciones para cada valor de lambda
for landa in landas:
    # Encuentra el eigenvector correspondiente al valor de lambda
    eigenvalue_index = np.argmin(np.abs(eigenvalues - landa))
    eigenvector = eigenvectors[:, eigenvalue_index]
    
    # Normaliza el eigenvector
    eigenvector_normalized = eigenvector / np.linalg.norm(eigenvector)
    
    print(f"Eigenvector correspondiente al eigenvalor {landa} encontrado:")
    print(eigenvector_normalized)
    
    # Calcula la proyección de los datos normalizados sobre el eigenvector
    proyeccion = df_normalizado.dot(eigenvector_normalized)
    print(f"La proyeccion para mi eigenvalue {landa} es:", proyeccion)
    
    # Reconstruye los datos originales utilizando la proyección y el eigenvector
    datos_reconstruidos = proyeccion.dot(eigenvector_normalized)
    print(f"Los datos reconstruidos para mi eigenvalue {landa} son:", datos_reconstruidos)
    
    # Calcula el error cuadrático medio (MSE) entre los datos originales y los datos reconstruidos
    mse = np.mean(np.square(df_normalizado.values - datos_reconstruidos))
    print("Error cuadrático medio (MSE):", mse)
```

```python
def pca(datos, num_componentes):
    # Paso 1: Normalización de los datos
    media = np.mean(datos, axis=0)
    datos_normalizados = datos - media
    std = np.std(datos_normalizados, axis=0)
    datos_normalizados /= std

    # Paso 2: Cálculo de la matriz de covarianza
    matriz_covarianza = np.cov(datos_normalizados, rowvar=False)
    print("Matriz de Covarianza:")
    print(matriz_covarianza)

    # Paso 3: Cálculo de los autovectores y autovalores
    autovalores, autovectores = np.linalg.eig(matriz_covarianza)
    print("Autovalores:")
    print(autovalores)

    # Paso 4: Selección de componentes principales
    indices_autovectores = np.argsort(autovalores)[::-1][:num_componentes]
    autovectores_principales = autovectores[:, indices_autovectores]

    # Paso 5: Proyección de los datos
    datos_proyectados = np.dot(datos_normalizados, autovectores_principales)

    # Paso 6: Reconstrucción de los datos (opcional)
    datos_reconstruidos = np.dot(datos_proyectados, autovectores_principales.T)

    return datos_proyectados, autovectores_principales, datos_reconstruidos
datos=df_sin_departamento
num_componentes=1
datos_proyectados, autovectores_principales, datos_reconstruidos = pca(datos, num_componentes)
print("Datos proyectados:")
print(datos_proyectados)
print("\nAutovectores principales:")
print(autovectores_principales)
print("\nDatos reconstruidos:")
print(datos_reconstruidos)
mse = np.mean(np.square(df_normalizado.values - datos_reconstruidos))
print("Error cuadrático medio (MSE):", mse)
```

```python
import os
import zipfile
from urllib.request import urlretrieve
from os.path import isfile, isdir

def download_files():
    """
    This method downloads image files if they do not exist.
    """
    try:
        data_dir = "data"
        if not isdir(data_dir):
            os.makedirs(data_dir)
        
        path_tar = os.path.join(data_dir, 'faces.zip')
        if not isfile(path_tar):
            urlretrieve('http://courses.media.mit.edu/2002fall/mas622j/proj/faces/rawdata.zip', path_tar)

        dest_path = r"data\faces"
        
        with zipfile.ZipFile(path_tar) as tar:
            tar.extractall(dest_path)
        
    except Exception as e:
        print("An error occurred:", e)

download_files()
```
